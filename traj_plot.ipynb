```python
import numpy as np
import matplotlib.pyplot as plt

# Plot each trajectory
plt.figure(figsize=(8, 6))
for i in range(10):
    data = np.load(f'./results/trajectory_TD3_64_128_2_{i}.npz')
    pose = data['Trajectory']
    x = pose[:, 0]  # Extract x coordinates
    y = pose[:, 1]  # Extract y coordinates
    plt.plot(x, y, label=f'Trajectory {i+1}', marker='o', markersize=5)  # Plot each trajectory

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Trajectory Plot')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
